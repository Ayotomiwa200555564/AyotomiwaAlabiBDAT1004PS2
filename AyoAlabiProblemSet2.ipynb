```python
#Question 1
a = 0 #variable a is set to 0
def b():
    global a #b is defined and makes use of the global variable a
    a = c(a) #a becomes c(a)
def c(a):
    return a + 2 #2 will be added to the value of a at this point after running
```

```python
>>> b()
>>> b()
>>> b()
>>> a
```

The value of "a" is modified every time the function b() calls the c(a).
In this case b() was called thrice
first instance: the return value of a = 0 + 2 = 2, the second time, a = 2 +2= 4, third time, a = 4+2= 6. 
This executes like an accumulator loop pattern

```python
#Question 2
#define the function 'fileLength' that takes the 'filename' as the argument
def fileLength(filename):
    try:
        infile = open(filename) #open the intended file 
        content = infile.read() #read the entire content of the file
        length = len(content)   #counts the length of the content
        print(length)
    except FileNotFoundError: #indicate the ExceptionType when using try-except statement
        print(f"File {filename} not found.")
```

```python
#Question 3
class Marsupial:
    def __init__(self):
        self.pouch = []

    def put_in_pouch(self, item):
        self.pouch.append(item)

    def pouch_contents(self):
        return self.pouch
    
class Kangaroo(Marsupial):
    def __init__(self,xcoord,ycoord):
    #initialize coordinates to xcoord,ycoord
        self.pouch = []
    #set x and y coordinates 
        self.x = xcoord
        self.y = ycoord
        
    def jump(self, dx, dy):
        self.x += dx
        self.y += dy 
        
    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"
    
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')

# Print the pouch contents separately
print(m.pouch_contents())

k = Kangaroo(0, 0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)
```

```python
#Question 4
def collatz(x):
    if x == 1:
        print(x)
        return
    # in cases where x>1,Print the current value of x
    print(x)
    if x % 2 == 0:
        collatz(x // 2)  # If x is even
    else:
        collatz(3 * x + 1)  # If x is odd
```

```python
#Question 6
infile = open("C:\\Users\\User\\OneDrive - Georgian College\\Documents\\BDAT 1004 Data Programming\\Problem Set 2\\w3c.txt", 'r')
content = infile.read()
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.in_heading = False
        self.indentation = 0

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h') and tag[1:].isdigit():
            self.in_heading = True
            self.indentation = int(tag[1:]) - 1  # Adjust indentation based on heading level

    def handle_data(self, data):
        if self.in_heading:
            indentation_spaces = '' * self.indentation
            print(f"{indentation_spaces}{data}")

    def handle_endtag(self, tag):
        if tag.startswith('h') and tag[1:].isdigit():
            self.in_heading = False

hp = HeadingParser()
hp.feed(content)
```

```python
#Question 7
import requests
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def __init__(self, depth, indent):
        super().__init__()
        self.depth = depth
        self.indent = indent

    def handle_starttag(self, tag, attrs):
        if self.depth >= 0:
            url = dict(attrs).get('href')
            if url and url.startswith('http'):
                print(' ' * self.indent + url)
                if self.depth > 0:
                    self.visit_url(url, self.depth - 1, self.indent + 4)
    def visit_url(self, url, depth, indent):
        try:
            response = requests.get(url)
            response.raise_for_status()
            parser = MyHTMLParser(depth, indent)
            parser.feed(response.text)
        except requests.exceptions.RequestException:
            pass

def webdir(url, depth, indent):
    try:
        response = requests.get(url)
        response.raise_for_status()
        parser = MyHTMLParser(depth, indent)
        parser.feed(response.text)
    except requests.exceptions.RequestException:
        pass

# example URL
webdir('http://web.simmons.edu/~grovesd/comm244/notes/week3/html-test-page.html', 2, 0)
```

```python
#Question 8
import sqlite3
conn = sqlite3.connect('newdata.db')
cursor = conn.cursor()

# Create a table to store the data
cursor.execute('''CREATE TABLE IF NOT EXISTS Weather (
    City TEXT,
    Country TEXT,
    Season TEXT,
    Temperature REAL,
    Rainfall REAL
)''')

data = [
    ('Mumbai', 'India', 'Winter', 24.8, 5.9),
    ('Mumbai', 'India', 'Spring', 28.4, 16.2),
    ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
    ('Mumbai', 'India', 'Fall', 27.6, 346.0),
    ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
    ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
    ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
    ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
    ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
    ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
    ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
    ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)
]

for row in data:
    cursor.execute('SELECT 1 FROM Weather WHERE City=? AND Country=? AND Season=?', (row[0], row[1], row[2]))
    if not cursor.fetchone():
        cursor.execute('INSERT INTO Weather VALUES (?, ?, ?, ?, ?)', row)

cursor.execute('SELECT * FROM Weather')
rows = cursor.fetchall()

# Print the selected data
for row in rows:
    print(row)

# Commit changes and close the connection
conn.commit()
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('newdata.db')
cursor = conn.cursor()

# a) All the temperature data
cursor.execute('SELECT Temperature FROM Weather')
temperature_data = cursor.fetchall()
print("a) All the temperature data:")
for temperature in temperature_data:
    print(temperature[0])

# b) All the cities, but without repetition
cursor.execute('SELECT DISTINCT City FROM Weather')
cities = cursor.fetchall()
print("\nb) All the cities, but without repetition:")
for city in cities:
    print(city[0])

# c) All the records for India
cursor.execute('SELECT * FROM Weather WHERE Country = "India"')
india_records = cursor.fetchall()
print("\nc) All the records for India:")
for record in india_records:
    print(record)

# d) All the Fall records
cursor.execute('SELECT * FROM Weather WHERE Season = "Fall"')
fall_records = cursor.fetchall()
print("\nd) All the Fall records:")
for record in fall_records:
    print(record)

# e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters
cursor.execute('''
    SELECT City, Country, Season
    FROM Weather
    WHERE Rainfall BETWEEN 200 AND 400
''')
rainfall_between_200_and_400 = cursor.fetchall()
print("\ne) The city, country, and season for which the average rainfall is between 200 and 400 millimeters:")
for record in rainfall_between_200_and_400:
    print(record)

# f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order
cursor.execute('''
    SELECT City, Country
    FROM Weather
    WHERE Season = "Fall" AND Temperature > 20
    ORDER BY Temperature
''')
above_20_degrees_increasing = cursor.fetchall()
print("\nf) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order:")
for record in above_20_degrees_increasing:
    print(record)

# g) The total annual rainfall for Cairo
cursor.execute('SELECT SUM(Rainfall) FROM Weather WHERE City = "Cairo"')
cairo_total_rainfall = cursor.fetchone()[0]
print("\ng) The total annual rainfall for Cairo:")
print(cairo_total_rainfall)

# h) The total rainfall for each season
cursor.execute('''
    SELECT Season, SUM(Rainfall)
    FROM Weather
    GROUP BY Season
''')
seasonal_rainfall = cursor.fetchall()
print("\nh) The total rainfall for each season:")
for record in seasonal_rainfall:
    print(record)

# Close the connection to the database
conn.close()
```

```python
#Question 9
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
#create empty lists to store each case
uppercase_words = []
lowercase_words = []
lengthofWords = []
wordDetails = []
selected_words =[]

for word in words:
    uppercase = word.upper()
    uppercase_words.append(uppercase)
    
    lowercase = word.lower()
    lowercase_words.append(lowercase)
   
    length = len(word)
    lengthofWords.append(length)
    
    details = [word.upper(), word.lower(), len(word)]
    wordDetails.append(details)
    if len(word)>=4:
        selected_word = word
        selected_words.append(selected_word)
        
print('a.', uppercase_words)
print()
print('b.', lowercase_words)
print()
print('c.', lengthofWords)
print()
print('d.', wordDetails)
print()
print('e.', selected_words)
```
